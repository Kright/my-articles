```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
def read_file(path: str):
    lines = Path(path).read_text().split("\n")
    nums = [[float(a) for a in l.split(",")] for l in lines]

    x = np.linspace(0, 100, num = len(lines))
    y0 = [i[0] for i in nums]
    y1 = [i[1] for i in nums]
    return x, y0, y1
```

```python
def show_solvers(name: str):
    fig, plots = plt.subplots(1, 3)
    for p, s in zip(plots, ["0.1", "0.01", "0.001"]):
        x, y0, y1 = read_file(f"csv/{name} {s}.csv")
        p.plot(x, y0)
        p.plot(x, y1)
```

```python
def show_solver(name: str, n: int = 3):
    s = ["1.0", "0.1", "0.01", "0.001", "1.0E-4", "1.0E-5"][n]
    x, y0, y1 = read_file(f"csv/{name} {s}.csv")
    plt.plot(x, y0)
    plt.plot(x, y1)
```

```python
plt.rcParams['figure.figsize'] = [20, 8]
```

```python
show_solver("SolverEulerNaive", n=2)
```
